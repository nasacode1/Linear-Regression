```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.random.rand(100, 1)
# y = 3*x + 1    y = wx+b
y = 3 * X.squeeze() + 1 + np.random.rand(100) * 0.3

plt.scatter(X, y, alpha=0.7)
plt.title("synthetic data")
plt.xlabel("x"); plt.ylabel("y")
plt.show()
```

```python
w = 0.0
b = 0.0
lr = 0.05
losses = []
for epoch in range(10000000):
  y_pred = w * X.squeeze() + b
  loss = np.mean((y - y_pred) ** 2)
  losses.append(loss)
  dw = -2 * np.mean((X.squeeze()*(y- y_pred)))
  db = -2 * np.mean(y - y_pred)

  w = w - lr * dw
  b = b - lr * b

print(f"w={w:.4f}, b = {b:.4f}, loss = {losses[-1]:.6f}")
```

```python
plt.scatter(X, y, alpha=0.7)
xs = np.linspace(0, 1, 100)
plt.plot(xs, w*xs+b, color='red')
plt.show()

plt.plot(losses)
plt.title("training loss over epochs")
plt.xlabel("epoch")
plt.ylabel("MSE")
plt.show()
```
